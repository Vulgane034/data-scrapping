INITIALISATION DES FONCTION DE NETTOYAGE

```python
# nettoyage et convertion des prix (debut de preparartion pour exploration de la DataFrame) 
def clean_price(price):
    if isinstance(price, str):
          price = price.replace("\xa0", "").replace("Â", "").replace("FCFA", "").replace(",", "").strip()
    return float(price) if price else None
    return price
# la fonction permet de remplacer \xa0 par des espace normal remplace FCFA par des espace et convertis le prix en float
```

```python
# nettoyage et convertion des reduction 
def clean_percentage(percentage):
    if isinstance(percentage, str):
        return float(percentage.replace("%", "").strip())
    return percentage
```

```python
# nettoyage des colonne prix reel et anciens prix 
df['prix reel'] = df['prix reel'].apply(clean_price)
df['anciens prix'] = df['anciens prix'].apply(clean_price)
```

```python
df['reduction'] = df['reduction'].apply(clean_percentage)
```

```python
def filter_by_category(df, category_word):
    # Vérifier si la colonne 'category' existe dans le DataFrame
    if 'category' not in df.columns:
        print("La colonne 'category' n'existe pas dans le DataFrame.")
        return
    
    # Utiliser la méthode query pour filtrer par la catégorie spécifiée
    filter_1 = df.query('category == @category_word')
    
    # Afficher le DataFrame filtré
    return filter_1    
```
